# That Looked Pretty Crazy
So let's actually turn it into something we can use - a shape.

````python
! pip install requests
"""
I have to first call the webservice to get the map area data for a specific coordinate. It'll be empty `{}` if there isn't a parcel there, 
i.e the Ocean, I picked a point not in the USA. 

"""
import requests

# --- Coordinates (longitude, latitude) ---
lon, lat = -122.449871, 37.492633

# --- SQL query ---
query = f"""
SELECT mup.mupolygonkey, mup.mukey, mup.mupolygongeo
FROM mupolygon AS mup
WHERE mup.mukey IN (
  SELECT mukey
  FROM SDA_Get_Mukey_from_intersection_with_WktWgs84('POINT({lon} {lat})')
)
"""

# --- POST request ---
url = "https://sdmdataaccess.nrcs.usda.gov/Tabular/post.rest"
response = requests.post(url, data={"query": query, "format": "json"}, headers={"Content-Type": "application/x-www-form-urlencoded"})
response.raise_for_status()

data = response.json()
````

````python
! pip install geopandas shapely
"""
After getting the data, I can now convert the data into a GeoDataFrame for further analysis or visualization.
"""
# --- Extract records ---
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Assuming your JSON is in a variable called `data`
rows = data["Table"]

# Create DataFrame
df = pd.DataFrame(rows, columns=["MUKEY", "AREASYMBOL", "geometry"])

# Convert WKT to actual polygons
df["geometry"] = df["geometry"].apply(wkt.loads)

# Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")

# Now you can plot or export
gdf.plot()
````

# Many Parcels
Even though this query runs fine, there may be dozens of polygons returned for a single point. This is because the Soil Webservice doesn't actually 
allow filtering to just the specific polygon that contains your point — only by the MUKEY. Not only that, but some MUKEYS may be repeated

![Parcels from the many MAPUNITS that belong to the coordinate (-122.449871, 37.492633) ](Working with Soil/assets/possible-parcels.png "The Many Parcels of a Single Coordinate")

Using a bit more effort, the actual polygon containing the coordinate can be found with just a few more lines of code to get a single parcel.

![The shapefile that contains the coordinate (-122.449871, 37.492633) ](Working with Soil/assets/found-parcel.png "The Parcel Containing the Coordinate")


Projecting the Parcels on a map can provide a greater sense of the underlying data.

![Parcels from the many MAPUNITS projected on a map.](Working with Soil/assets/parcel-projection-on-map.png "The Many Parcels of a Single Coordinate")


````python
"""
Notice here there's many points! In order to get the point the coordinate belongs to,
check each polygon.
"""
from shapely import Point
import geopandas as gpd
import pandas as pd

lon, lat = -122.449871, 37.492633
pt = Point(lon, lat)

gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")

# Filter to the polygon(s) containing your point
containing = gdf[gdf.contains(pt)]

print(containing)
containing.plot()
````

````python
! pip install folium
"""
Finally, I can visualize the polygons on an interactive map using Folium.
"""
import folium
from shapely import wkt
import json

rows = data["Table"]
polygons = [wkt.loads(r[2]) for r in rows]

# Create a Folium map centered roughly on your polygon
center = polygons[0].centroid.coords[0][::-1]  # (lat, lon)
m = folium.Map(location=center, zoom_start=10, tiles="CartoDB positron")

# Add polygons
for poly in polygons:
    geojson = json.loads(gpd.GeoSeries([poly]).to_json())
    folium.GeoJson(geojson, style_function=lambda x: {
        "fillColor": "orange",
        "color": "black",
        "weight": 1,
        "fillOpacity": 0.5
    }).add_to(m)

m.save("soil_polygon_map.html")
m
````
